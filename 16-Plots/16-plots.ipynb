Histogram plot:
Now, let’s generate a histogram for the df. Histogram plot can be created with DataFrame.plot.hist() function.

df.plot.hist()

When creating a histogram using the `plt.hist()` function in pandas (which is often used in conjunction with Matplotlib), you can customize various parameters to control the appearance and behavior of the histogram. Here are some of the most commonly used parameters:

1. `x`: This parameter specifies the data that you want to create a histogram for. It can be a Pandas Series, a list, or an array.

2. `bins`: This parameter determines the number of bins or intervals into which your data will be divided. You can specify it as an integer or a sequence of bin edges. For example, `bins=10` will create 10 equally spaced bins.

3. `range`: You can specify the range of values to include in the histogram using this parameter. It's a tuple that defines the minimum and maximum values for the bins. For example, `range=(0, 100)` restricts the histogram to values between 0 and 100.

4. `density`: When set to `True`, the histogram will display the density instead of the raw counts. The area under the histogram will sum to 1 when using density.

5. `cumulative`: If set to `True`, the histogram will be cumulative, showing the cumulative counts or density.

6. `color`: You can specify the color of the bars in the histogram. For example, `color='skyblue'`.

7. `edgecolor`: This parameter sets the color of the edges of the bars in the histogram.

8. `alpha`: It controls the transparency of the bars. A value between 0 (completely transparent) and 1 (completely opaque) can be used.

9. `histtype`: Specifies the type of histogram. Options include 'bar' (the default), 'barstacked' (for stacked bars), 'step' (step histogram), and 'stepfilled' (filled step histogram).

10. `orientation`: You can set the orientation of the histogram to 'horizontal' by setting `orientation='horizontal'`.

11. `label`: This parameter allows you to specify a label for the histogram, which can be used in legends.

12. `ax`: You can specify the axes on which to draw the histogram if you have multiple subplots.

Here's an example of how you can use some of these parameters:


```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.Series([87, 23, 45, 65, 32, 43, 12, 78, 89, 67, 54, 34, 65, 76, 34, 54, 23, 45, 67, 12, 87, 65, 34, 23, 43, 54, 65, 76, 45, 12])

# Create a histogram plot
data.plot.hist(bins=5, edgecolor='black', )  # You can adjust the number of bins for the histogram
```

```python
data = pd.Series([87, 23, 45, 65, 32, 43, 12, 78, 89, 67, 54, 34, 65, 76, 34, 54, 23, 45, 67, 12, 87, 65, 34, 23, 43, 54, 65, 76, 45, 12])

plt.hist(data, bins=10, range=(0, 100), density=True, color='skyblue', edgecolor='black', alpha=0.7, histtype='bar', label='Data Distribution')

plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Histogram Example')
plt.legend()

plt.show()
```

### Box plot:
Let's 
 create box plot. Box plot can be created with `DataFrame.plot.box()` function or `DataFrame.boxplot()`.

 ![Box Plot](box-plot.png)

```python
#generating the box plot in a horizontal form.

df.plot.box(vert=False, positions=[1, 2, 3, 4]);
```

**Attributes**

1. `column`: This parameter specifies which column(s) of the DataFrame you want to create box plots for. You can provide a single column name or a list of column names.

2. `by`: If you have a DataFrame with multiple groups and you want to create separate box plots for each group, you can use the `by` parameter. It specifies a column by which the data will be grouped, and separate box plots will be created for each group.

3. `ax`: You can specify the Matplotlib axes object where the box plot will be drawn. This allows you to create subplots or combine multiple box plots on the same figure.

4. `vert`: By default, box plots are vertical (i.e., boxes are oriented vertically). You can set `vert=False` to create horizontal box plots.

5. `patch_artist`: When set to `True`, the individual elements of the box plot (boxes, whiskers, etc.) are drawn as "artists," allowing you to customize their appearance (colors, fill, etc.).

6. `showfliers`: You can choose whether to display outliers (data points that fall outside the whiskers) by setting `showfliers` to `True` or `False`.

7. `notch`: When set to `True`, a notch is added to the box plot to indicate the confidence interval around the median.

8. `whis`: This parameter determines the length of the whiskers as a factor of the interquartile range (IQR). By default, it's set to 1.5, but you can customize it as needed.

9. `meanline`: Set this parameter to `True` to add a line indicating the mean value of the data within each box.

10. `showmeans`: When set to `True`, a marker is added to the mean line to indicate the mean value.

11. `meanprops`: You can use this parameter to customize the appearance of the mean marker.

12. `medianprops`: Similarly, you can use this parameter to customize the appearance of the median line.

13. `boxprops`: Customize the appearance of the box (the rectangle) in the box plot.

14. `whiskerprops`: Customize the appearance of the whiskers.

15. `capprops`: Customize the appearance of the caps (the ends of the whiskers).

16. `bootstrap`: By default, the confidence intervals around the median are computed using bootstrapping. You can set `bootstrap` to `None` to disable bootstrapping or provide a specific number of bootstrapping iterations.


```python
data = {
    'Group A': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    'Group B': [40, 45, 50, 55, 60, 65, 70, 75, 80, 85]
}

dfnew = pd.DataFrame(data)

# Create a box plot with custom parameters
dfnew.boxplot(column=['Group A', 'Group B'], notch=True, showfliers=True, patch_artist=True, meanline=True)

plt.title('Customized Box Plot')
plt.ylabel('Values')
plt.xlabel('Groups')
plt.grid(True)

plt.show()
```

### Scatter plot:

A Scatter plot can be created with DataFrame.plot.scatter() function. As we know scatter plot takes two-positional required arguments i.e. x and y to plot the graph. So, we will give the values of the  x and y-axis as the name of columns

```python
df2 = pd.DataFrame(data = np.random.randint(1, 20, 15), columns = ["value"])
df2["index"]= df2.index
print(df2) 
df2.plot.scatter(x="index", y="value", c="value" )
```

2. **Markers**: You can specify the markers used for individual data points on the scatter plot. Common markers include circles, squares, triangles, and more. You can customize the marker type, size, color, and style.

3. **Colors**: You can set custom colors for the markers to distinguish different groups or data points. Colors can be specified as strings (e.g., 'red', 'blue') or RGB values.

4. **Alpha**: The alpha parameter controls the transparency of the markers. It accepts values between 0 (completely transparent) and 1 (completely opaque).

5. **Sizes**: You can adjust the size of the markers to emphasize or deemphasize individual data points. Sizes can be fixed or vary based on a data variable.

6. **Labels**: If you want to label specific data points or group them into categories, you can add labels to the markers.

7. **Legend**: When plotting multiple groups of data on the same scatter plot, a legend can be added to identify each group.

8. **Title and Labels**: You can provide titles for the plot and labels for the x and y axes to provide context and information.

9. **Grid**: You can add grid lines to the plot to aid in data interpretation.

10. **Axis Limits**: Customize the range displayed on the x and y axes by setting axis limits.

11. **Color Maps (Colormaps)**: When you have a third variable that you want to map to marker colors, you can use colormaps to create color gradients.

12. **Regression Lines**: If you're interested in fitting a regression line to the data, you can add a regression line to visualize the linear relationship.

### Pie chart:

A Pie plot can be created with DataFrame.plot.pie() function or Series.plot.pie(). To generate a pie chart we will create series data as a pie chart is created only for one column. Let’s create a series named pie.

```python
pie = pd.Series(np.random.rand(5))
pie.plot.pie();
```

```python
df.plot.pie(subplots=True, figsize=(15, 15))
```

When creating a pie chart in Python using libraries like Matplotlib, you can customize its appearance and behavior by using various attributes. Below are some common attributes and parameters you can use when working with pie charts:

1. **Labels**: You can provide labels for each slice (or wedge) of the pie chart. These labels typically describe what each slice represents.

2. **Sizes**: Pie charts represent data as proportions of the whole, so you need to provide the sizes or values for each slice. These sizes should add up to 100% or 1 (if expressed as fractions).

3. **Colors**: You can set custom colors for each slice to make your pie chart more visually appealing or to emphasize specific data points.

4. **Explode**: You can use the `explode` parameter to emphasize a particular slice by "exploding" it, which means pulling it out from the center of the chart. This can be useful for highlighting specific data points.

5. **Start Angle**: By default, pie charts start drawing slices from the positive x-axis (3 o'clock position) and move counterclockwise. You can specify a custom start angle using the `startangle` parameter.

6. **Shadow**: If you want to add a shadow effect to your pie chart, you can use the `shadow` parameter.

7. **Labels Distance**: The `labeldistance` parameter allows you to control the distance of the labels from the center of the pie chart.

8. **Autopct**: You can display the percentage of each slice by using the `autopct` parameter to specify the format of the percentage labels.

9. **Counterclock**: If you want the slices to be drawn clockwise instead of the default counterclockwise, you can use the `counterclock` parameter.

10. **Radius**: You can adjust the radius of the pie chart using the `radius` parameter.


```python
import matplotlib.pyplot as plt

# Data for the pie chart
sizes = [30, 20, 15, 35]  # Sizes or values for each slice
labels = ['Category A', 'Category B', 'Category C', 'Category D']  # Labels for each slice
colors = ['blue', 'green', 'red', 'purple']  # Custom colors for each slice
explode = (0.1, 0, 0, 0)  # "Explode" the first slice

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode, startangle=90, shadow=True, autopct='%1.1f%%')

# Add a title
plt.title('Sample Pie Chart')

# Display the chart
plt.show()
```
